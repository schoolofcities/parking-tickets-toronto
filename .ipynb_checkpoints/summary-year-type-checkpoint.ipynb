## Summarizing parking tickets by year and infraction type

```python
import pandas as pd
import os
import zipfile
```

```python
zf = zipfile.ZipFile("data/parking-tickets/parking-tickets-2019.zip")
csvs = zf.namelist()
csvs
```

```python
dfs = []
for csv in csvs:
    dfs.append(pd.read_csv(zf.open(csv)))
df = pd.concat(dfs)
```

```python
pd.DataFrame(df.value_counts(["infraction_description"]))
```
